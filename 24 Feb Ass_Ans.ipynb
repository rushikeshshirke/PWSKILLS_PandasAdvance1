```python
# Q2.
import pandas as pd

def reindex_dataframe(df):
    # Reset the existing index and create a new index starting from 1, incrementing by 2
    new_index = pd.Index(range(1, len(df)*2, 2), name='NewIndex')

    # Use set_index to re-index the DataFrame
    df_reindexed = df.set_index(new_index)

    return df_reindexed

# Example usage:
# Assuming df is your existing DataFrame with columns 'A', 'B', and 'C'
data = {'A': [10, 20, 30],
        'B': [40, 50, 60],
        'C': [70, 80, 90]}

df = pd.DataFrame(data)

# Call the function to re-index the DataFrame
df_reindexed = reindex_dataframe(df)

# Display the re-indexed DataFrame
print(df_reindexed)
```

```python
import pandas as pd

def calculate_sum_of_first_three(df):
    # Check if the 'Values' column exists in the DataFrame
    if 'Values' not in df.columns:
        print("Error: 'Values' column not found in the DataFrame.")
        return

    # Extract the 'Values' column
    values_column = df['Values']

    # Check if there are at least three values in the column
    if len(values_column) < 3:
        print("Error: Insufficient values in the 'Values' column.")
        return

    # Calculate the sum of the first three values
    sum_of_first_three = values_column.head(3).sum()

    # Print the sum to the console
    print("Sum of the first three values in 'Values' column:", sum_of_first_three)

# Example usage:
# Assuming df is your existing DataFrame with a column named 'Values'
data = {'Values': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Call the function to calculate and print the sum of the first three values
calculate_sum_of_first_three(df)
```

```python
# Q4.
import pandas as pd

def add_word_count_column(df):
    # Check if the 'Text' column exists in the DataFrame
    if 'Text' not in df.columns:
        print("Error: 'Text' column not found in the DataFrame.")
        return df

    # Calculate word count for each row in the 'Text' column
    df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()))

    return df

# Example usage:
# Assuming df is your existing DataFrame with a column named 'Text'
data = {'Text': ['This is a sample text.', 'Another example', 'One more sentence.']}

df = pd.DataFrame(data)

# Call the function to add the 'Word_Count' column
df = add_word_count_column(df)

# Display the DataFrame with the new 'Word_Count' column
print(df)
```

```python
# Q7.

import pandas as pd

def extract_username_from_email(df):
    # Check if the 'Email' column exists in the DataFrame
    if 'Email' not in df.columns:
        print("Error: 'Email' column not found in the DataFrame.")
        return df

    # Extract the username from each email address
    df['Username'] = df['Email'].apply(lambda x: x.split('@')[0] if '@' in x else None)

    return df

# Example usage:
# Assuming df is your existing DataFrame with a column named 'Email'
data = {'Email': ['john.doe@example.com', 'alice.smith@example.com', 'bob@example.com']}

df = pd.DataFrame(data)

# Call the function to add the 'Username' column
df = extract_username_from_email(df)

# Display the DataFrame with the new 'Username' column
print(df)
```

```python
# Q8.

import pandas as pd

def filter_dataframe(df):
    # Check if columns 'A' and 'B' exist in the DataFrame
    if 'A' not in df.columns or 'B' not in df.columns:
        print("Error: Columns 'A' and 'B' not found in the DataFrame.")
        return pd.DataFrame()

    # Use boolean indexing to select rows based on conditions
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]

    return selected_rows

# Example usage:
# Assuming df is your existing DataFrame with columns 'A', 'B', and 'C'
data = {'A': [3, 8, 6, 2, 9],
        'B': [5, 2, 9, 3, 1],
        'C': [1, 7, 4, 5, 2]}

df = pd.DataFrame(data)

# Call the function to filter the DataFrame
filtered_df = filter_dataframe(df)

# Display the new DataFrame containing selected rows
print(filtered_df)
```

```python
# Q9.

import pandas as pd

def calculate_statistics(df):
    # Check if the 'Values' column exists in the DataFrame
    if 'Values' not in df.columns:
        print("Error: 'Values' column not found in the DataFrame.")
        return None

    # Calculate mean, median, and standard deviation
    mean_value = df['Values'].mean()
    median_value = df['Values'].median()
    std_deviation = df['Values'].std()

    # Create a dictionary to store the results
    statistics_result = {'Mean': mean_value, 'Median': median_value, 'Std Deviation': std_deviation}

    return statistics_result

# Example usage:
# Assuming df is your existing DataFrame with a column named 'Values'
data = {'Values': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Call the function to calculate statistics
result = calculate_statistics(df)

# Display the results
print("Mean:", result['Mean'])
print("Median:", result['Median'])
print("Standard Deviation:", result['Std Deviation'])
```

```python
# Q10.

import pandas as pd

def calculate_moving_average(df):
    # Check if the 'Sales' and 'Date' columns exist in the DataFrame
    if 'Sales' not in df.columns or 'Date' not in df.columns:
        print("Error: 'Sales' or 'Date' column not found in the DataFrame.")
        return df

    # Convert the 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Sort the DataFrame by 'Date'
    df = df.sort_values(by='Date')

    # Calculate the moving average with a window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df

# Example usage:
# Assuming df is your existing DataFrame with columns 'Sales' and 'Date'
data = {'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05', '2022-01-06', '2022-01-07'],
        'Sales': [10, 15, 20, 25, 30, 35, 40]}

df = pd.DataFrame(data)

# Call the function to calculate and add the 'MovingAverage' column
df = calculate_moving_average(df)

# Display the DataFrame with the new 'MovingAverage' column
print(df)
```

```python
# Q11.

import pandas as pd

def add_weekday_column(df):
    # Check if the 'Date' column exists in the DataFrame
    if 'Date' not in df.columns:
        print("Error: 'Date' column not found in the DataFrame.")
        return df

    # Convert the 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Add a new 'Weekday' column with weekday names
    df['Weekday'] = df['Date'].dt.day_name()

    return df

# Example usage:
# Assuming df is your existing DataFrame with a column named 'Date'
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}

df = pd.DataFrame(data)

# Call the function to add the 'Weekday' column
df = add_weekday_column(df)

# Display the DataFrame with the new 'Weekday' column
print(df)
```

```python
# Q12.

import pandas as pd

def filter_date_range(df):
    # Check if the 'Date' column exists in the DataFrame
    if 'Date' not in df.columns:
        print("Error: 'Date' column not found in the DataFrame.")
        return df

    # Convert the 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Specify the date range
    start_date = pd.to_datetime('2023-01-01')
    end_date = pd.to_datetime('2023-01-31')

    # Use boolean indexing to select rows within the date range
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

    return selected_rows

# Example usage:
# Assuming df is your existing DataFrame with a column named 'Date'
data = {'Date': ['2023-01-01', '2023-01-15', '2023-01-25', '2023-02-05', '2023-02-15']}

df = pd.DataFrame(data)

# Call the function to filter the DataFrame based on the date range
filtered_df = filter_date_range(df)

# Display the new DataFrame containing selected rows
print(filtered_df)
```
